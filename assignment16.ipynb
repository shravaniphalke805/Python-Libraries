```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# List of numerical columns
numerical_columns = ['age', 'bmi', 'children', 'expenses']
```

```python
# Create distribution plots
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
# List of categorical columns
categorical_columns = ['sex', 'smoker', 'region']
```

```python
# Create count plots
for column in categorical_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=column, data=df, hue=column, palette='Set2', legend=False)
    plt.title(f'Count Plot of {column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Count')

    plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df[numerical_columns].corr()
```

```python

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5)

plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# 1. Find the average insurance charges
average_charges = df['expenses'].mean()
print("Average Insurance Charges:", average_charges)
```

```python
# 2. Find maximum and minimum charges
maximum_expenses = df['expenses'].max()
minimum_expenses = df['expenses'].min()
print("Maximum Insurance Expenses :", maximum_expenses)
print("Minimum Insurance Expenses:", minimum_expenses)
```

```python
# 3. Compare average charges for smokers vs non-smokers
avg_expenses_smoker = df.groupby('smoker')['expenses'].mean()

print("\nAverage Expenses by Smoking Status:")
print(avg_expenses_smoker)
```

```python
# Create boxplot
plt.figure(figsize=(8, 5))

sns.boxplot(
    x='smoker',
    y='expenses',
    hue='sex',
    data=df,
    palette='Set2'
)

plt.title('Distribution of Insurance Charges by Smoker Status and Sex')
plt.xlabel('Smoker Status')
plt.ylabel('Insurance Charges')

plt.show()
```
