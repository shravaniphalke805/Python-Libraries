```python
plt.figure(figsize=(10,5))
lowest_cartoon = df.nsmallest(10, 'Rating')
sns.barplot(    x='Rating',
    y='Name',
    data=lowest_cartoon,
    palette='viridis',
    hue='Name',
    legend=False)
plt.title("10 Lowest Rated Cartoons")
plt.xlabel("Rating")
plt.ylabel("Cartoon Name")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Rating'],
    bins=10,
    kde=True,
    color='orange'
)

plt.title("Distribution of Cartoon Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Cartoons")
plt.show()
```

```python
df['Cartoon_Index'] = range(1, len(df)+1)
plt.figure(figsize=(10,5))
sns.scatterplot(
    x='Cartoon_Index',
    y='Rating',
    data=df
)
plt.title("Scatter Plot of Cartoon Index vs Rating")
plt.xlabel("Cartoon Index")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    y=df['Rating'],
    color='skyblue'
)

plt.title("Box Plot of Cartoon Ratings")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,6))

df['Rating_Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['Lavender','Pink','lightblue']
)

plt.title("Cartoon Rating Categories")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='Name',
    data=df.head(20),
    palette='coolwarm',
    hue='Name',
    legend=False
)

plt.title("Cartoon Name Distribution (First 20 Records)")
plt.xlabel("Count")
plt.ylabel("Cartoon Name")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(
    df[['Rating','Start_Year']].corr(),
    annot=True,
    cmap='coolwarm')
plt.title("Correlation Between Cartoon Rating and Release Year")
plt.show()
```
